```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import mean_absolute_error
from xgboost import XGBRegressor
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("ford.csv")
```

```python
data.head()
```

```python
data
```

```python
data.isnull().sum()
```

```python
print(data["transmission"].unique(),
data["fuelType"].unique())
```

```python
data.replace({"transmission":{"Automatic": 0, "Manual": 1, "Semi-Auto": 2}}, inplace=True)
```

```python
data.replace({"fuelType":{"Petrol": 0, "Diesel": 1, "Hybrid": 2, "Electric": 3, "Other": 4}}, inplace=True)
```

```python
data
```

```python
x = data.drop(["model","price"], axis=1).values
y = data["price"]
```

```python
y
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
strad_c = scaler.transform(x)
scaler.transform(x)
```

```python
x = scaler.transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state = 42)
print(x.shape, x_train.shape, x_test.shape)
print(y.shape, y_train.shape, y_test.shape)
```

```python
model = XGBRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error

training_data_pred = model.predict(x_train)

score1 = r2_score(y_train, training_data_pred)

mae = mean_absolute_error(y_train, training_data_pred)

print("R2 score on training data: ", score1)
print("Mean Absolute Error on training data: ", mae)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error

test_data_pred = model.predict(x_test)

score1 = r2_score(y_test, test_data_pred)

mae = mean_absolute_error(y_test, test_data_pred)

print("R2 score on test data: ", score1)
print("Mean Absolute Error on test data: ", mae)
```

```python
input_data = (2018, 0,	1482,	0,	145,	48.7,	1.0)

input_changed = np.array(input_data).reshape(1,-1)

std_input = scaler.transform(input_changed)

prediction = model.predict(std_input)

print("this car estimation is: ", prediction)
```
